```python
import matplotlib.pyplot as plt
```

```python
#initialization of data

x = [1,2,3,4,5]
y = [63,75,77,84,95]
```

```python
#ploting the data 

plt.plot(x,y)
```

```python
# adding the values

x = [5,7,3,9,6]
y = [73,55,69,92,66]

#adding the Title
plt.title("report")

#adding the label 
plt.xlabel("rank.")
plt.ylabel("% of marks")
plt.plot(x,y)
plt.show()

```

```python
#   BAR  CHART

plt.title("report")

#adding the label 
plt.xlabel("semester no.")
plt.ylabel("% of marks")
plt.bar(x,y)
plt.show()

```

```python
#  scatter plot

# adding the values

x = [5,7,3,9,6]
y = [73,55,69,92,66]

#adding the Title
plt.title("report")

#adding the label 
plt.xlabel("rank.")
plt.ylabel("% of marks")
plt.scatter(x,y)
plt.show()
```

```python
# histogram chart


# adding the values

x = [1,2,3,4,5]
plt.hist(x,bins =[1,2,3,4,5])
#adding the Title
plt.title("report")

#adding the label 
plt.xlabel("rank.")
plt.ylabel("% of marks")
plt.legend(["bar"])
plt.show()
```

```python
#  pie chart 

#adding the values
#data to display on plot
x=[56,43,88,4]
e = (0,0,0,0.1)
plt.pie(x, explode= e)
plt.title("PIE")
plt.show()
```
