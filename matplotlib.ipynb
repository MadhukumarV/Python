# MATPLOT LIB


Matplotlib is a low level graph plotting library in python that serves as a visualization utility.








```python
import matplotlib
```

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x= np.array([1,9,4,9])
y=np.array([2,6,8,3])
```

```python
x
```

```python
plt.plot(x,y)
plt.show()
```

MARKER 

```python
plt.plot(x,y,c='#4dffc9',linewidth='10',ls='--')    #marker gives the line with *
plt.show()                                           # c= color,ls=line style
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y,c='#FBBC05')
plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse",c='#4dffc9')
plt.ylabel("Calorie Burnage",c='#4dffc9')

plt.show()
```

```python

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
```

```python
plt.plot(x,y,marker='H',ls='-.')

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```
